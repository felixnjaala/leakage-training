```python
import matplotlib.pyplot as plt
G7_brothers=["felly", "chris","jongela","sinbady","mataya","chussi","auba"]
G7_salary=[35,56,84,45,69,90,70]
plt.figure(figsize=(10,10))
plt.pie(G7_salary,labels=G7_brothers,startangle=90,shadow=True,explode=[0,0,0,0,0,0.07,0],colors=["blue","red","green","yellow","pink","orange","green"],autopct="%2.0f%%")
plt.legend(title="G7_brothers")
plt.show()
```
