Seaborn barplot it include one or more categorical values with a single numerical value

```python
#bar plot
plt.figure(figsize=(8,6))
sns.barplot(x='SaleCondition', y='SalePrice', data=bg,ci=None, estimator=np.std,palette='viridis' );
plt.title('Barplot')
plt.xlabel('categorical values')
plt.ylabel('numerical value')
plt.show()
```

```python
#histogram
plt.figure(figsize=(12,6))
sns.histplot(x='LotFrontage', data=bg)
```

```python
#scatter plot
sns.scatterplot(data=bg, x='LotArea', y='MSSubClass',hue='Utilities', sizes=(20,200))
```

```python
#lineplot
sns.lineplot(data=bg, x='LotArea', y='SalePrice')
```

```python
#heatmap
cor = bg.corr(numeric_only=True)
plt.figure(figsize=(30,25))
sns.heatmap(data=cor, annot=True,cmap='coolwarm', linewidths=0.6)
```
